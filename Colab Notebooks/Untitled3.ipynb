```python
import torch

param = torch.tensor([0.22, 3.4, -0.43, 4.7])
scale = param.max() - param.min()
# print(scale)
bits = torch.tensor([8, 8, 8, 8])
unit = 1 / (2**bits - 1)
noise_source = torch.randn_like(param) / 2
noise = scale * unit * noise_source
noisy = param + noise
print(noise)
print(noisy)
```
