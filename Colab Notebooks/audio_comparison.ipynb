```python
import tensorflow as tf

first_audio_path = "/content/jahaje.wav"
second_audio_path = "/content/jahaje.wav"

raw_audio = tf.io.read_file(first_audio_path)
first_audio, sample_rate = tf.audio.decode_wav( raw_audio, desired_channels=1)

first_audio = tf.reshape(first_audio,[-1])

raw_audio = tf.io.read_file(second_audio_path)
second_audio, sample_rate = tf.audio.decode_wav( raw_audio, desired_channels=1)
second_audio = tf.reshape(second_audio,[-1])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import librosa
import IPython.display as ipd
plt.figure(figsize=(14, 5))
librosa.display.waveplot(first_audio.numpy(), sr=48000)
librosa.display.waveplot(second_audio.numpy(), sr=48000)
```

```python
error = 0
for (val, val2) in zip(first_audio, second_audio):
  error += (tf.get_static_value((val - val2)))

print(error)
```
